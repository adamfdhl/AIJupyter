## 1. Membaca dataset standar iris dan dataset play-tennis

```python
from sklearn import datasets, tree
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import graphviz
```

Read Iris Dataset

```python
iris = datasets.load_iris()
iris_data = iris.data
iris_target = iris.target
# len(iris_data)
# len(iris_target)
# iris.feature_names
iris_data_frame = pd.DataFrame(
    data=np.c_[iris_data, iris_target], 
    columns=iris.feature_names + ['target'])
iris_data_frame.head()
```

```python
iris_data_frame.describe()
```

## 2. Melakukan pembelajaran dengan skema full training & tampilkan model

2.1 NaiveBayes

```python
gnb = GaussianNB()
gnb.fit(iris.data, iris.target)
```

```python
print("Probabilistic Model")
print("-------------------------------------")
print(gnb.class_prior_)
print("-------------------------------------")
print(gnb.theta_)
print("-------------------------------------")
print(gnb.sigma_)
```

2.3 DecisionTree ID3

```python
clf = tree.DecisionTreeClassifier()
clf.fit(iris_data, iris_target)
dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=iris.feature_names, 
                                class_names=iris.target_names, 
                                filled=True, rounded=True, 
                                special_characters=True)  
graph = graphviz.Source(dot_data)
# graph
```

2.4 kNN

```python
from sklearn.neighbors import KNeighborsClassifier

n = 5
knn = KNeighborsClassifier(n)
knn.fit(iris_data, iris_target)
```

2.5 Neural Network MLP

```python
from sklearn.neural_network import MLPClassifier
nnmlp = MLPClassifier()
nnmlp.fit(iris_data, iris_target)
```

```python
print("ANN Model")
print(nnmlp.coefs_)
print(nnmlp.intercepts_)
```
